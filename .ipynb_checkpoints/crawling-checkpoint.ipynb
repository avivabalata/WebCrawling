```python
# import the library used to query a website
from urllib.request import urlopen

# specify the url
wiki = "https://en.wikipedia.org/wiki/Gal_Gadot"
```

```python
# Query the website and return the html to the variable 'page'
page = urlopen(wiki)
```

```python
# import the Beautiful soup functions to parse the data returned from the website
from bs4 import BeautifulSoup
```

```python
#Parse the html in the 'page' variable, and store it in Beautiful Soup format
soup = BeautifulSoup(page)
```

## The html

```python
print(soup.prettify())
```

```python

films_table = soup.find("table",class_="wikitable sortable")

print(films_table.prettify())
```

```python
#Generate lists
#Generate lists
A=[]
B=[]
C=[]
D=[]
rowSpanYear = 0
rowSpanYearValue = ""
rowSpanRole = 0
rowSpanRoleValue = ""

for row in films_table.tbody.findAll("tr"):
    cells = row.findAll('td')
    if len(cells) > 0:
        
        ##############
        
        if cells[0].has_attr('rowspan'):
            rowSpanYear = int(cells[0].attrs['rowspan'])
            rowSpanYearValue = cells[0].find(text=True)
        else:
            rowSpanYear = 0
    
        if rowSpanYear > 0: 
            A.append(rowSpanYearValue)
            rowSpanYear = - rowSpanYear
        else:
            A.append(cells[0].find(text=True))
            
        ############
        
        if cells[2].has_attr('rowspan'):
            rowSpanRole = int(cells[2].attrs['rowspan'])
            rowSpanRoleValue = cells[2].find(text=True)
        else:
            rowSpanRole = 0
    
        if rowSpanRole > 0: 
            C.append(rowSpanRoleValue)
            rowSpanRole = - rowSpanRole
        else:
            C.append(cells[2].find(text=True))
            
        #############
        
        if cells[1].find(text=True) != '':
            B.append(cells[1].find(text=True))
        if cells[3] & cells[3].find(text=True) != '':
            D.append(cells[3].find(text=True))
```

```python
#import pandas to convert list to data frame
import pandas as pd
df=pd.DataFrame()
df['Year']=A
df['Title']=B
df['Role']=C
df['Director']=D
df
```
